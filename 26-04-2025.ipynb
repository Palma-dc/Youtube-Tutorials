```python
from statsbombpy import sb
import pandas as pd
from mplsoccer import VerticalPitch,Pitch
```

```python
sb.competitions()
```

```python
events_df = sb.matches(competition_id=9, season_id=281)
```

```python
frames_df
```

```python
frames_df = sb.competition_frames(
country="Germany",
division="1. Bundesliga",
season="2023/2024",
gender="male")

frames_df.rename(columns={'event_uuid': 'id'}, inplace = True)
merged_df=pd.merge(frames_df, events_df,
how="left", on=["match_id","id"])
```

```python
df = merged_df.copy()
```

```python
df = df[(df['player'] == 'Florian Wirtz')&(df['type']== 'Shot')&(df['shot_outcome']== 'Goal')].reset_index(drop=True)
```

```python
df[['x_start','y_start']] = pd.DataFrame(df['location_y'].tolist(), index=df.index)
df[['x_end','y_end', 'z_end']] = pd.DataFrame(df['shot_end_location'].tolist(), index=df.index)
```

```python
p = VerticalPitch(pitch_color='#aabb97', line_color='white',
              stripe_color='#c2d59d', stripe=True, half = True)
fig, ax = p.draw()

df = df[0:1].reset_index(drop=True) 

p.scatter(x=df['x_start'], y=df['y_start'], ax=ax, color='black', edgecolor='red', s=100, lw=1.8)
p.lines(xstart=df['x_start'], ystart=df['y_start'], xend=df['x_end'], yend = df['y_end'],lw=2,  color='black',transparent=True,  ax=ax, comet=True) 

for x in df['shot_freeze_frame'][0]:
    if x['teammate']:
        color = 'black',
        edgecolor = 'red'
    else:
        if x['position']['id'] ==1:
            color = 'yellow'
            edgecolor = 'yellow'
        else:
            color = 'white',
            edgecolor = 'red'
    p.scatter(x=x['location'][0], y=x['location'][1], ax=ax, color=color, edgecolor=edgecolor, lw=1.5)
```

```python
fig.savefig('my_pitch_plot.png', dpi=300, bbox_inches='tight')
```
